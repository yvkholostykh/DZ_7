```python
import requests  # Импортируем библиотеку для работы с HTTP‑запросами

def fetch_and_display_posts():
    """
    Функция для получения и отображения заголовков и тел первых 5 постов
    из публичного API JSONPlaceholder.
    """
    # URL API для получения списка постов
    api_url = "https://jsonplaceholder.typicode.com/posts"

    try:
        # Отправляем GET‑запрос к API
        print("Отправка запроса к API...")
        response = requests.get(api_url)

        # Проверяем статус ответа (200 означает успешный запрос)
        if response.status_code == 200:
            print("Запрос успешен! Получаем данные...")

            # Парсим JSON‑ответ в список словарей
            posts = response.json()

            # Выводим заголовки и тела первых 5 постов
            print("\nПервые 5 постов:\n")
            for i, post in enumerate(posts[:5], start=1):
                # Извлекаем заголовок и тело поста
                title = post.get("title", "Заголовок не найден")
                body = post.get("body", "Тело поста не найдено")

                # Форматируем и выводим информацию
                print(f"Пост #{i}:")
                print(f"  Заголовок: {title}")
                print(f"  Тело: {body}")
                print("-" * 50)  # Разделитель между постами
        else:
            # Если статус ответа не 200, выводим ошибку
            print(f"Ошибка: получен статус {response.status_code}")

    except requests.exceptions.RequestException as e:
        # Обрабатываем возможные ошибки сети (нет соединения, тайм‑аут и т. д.)
        print(f"Произошла ошибка при выполнении запроса: {e}")
    except ValueError as e:
        # Обрабатываем ошибки парсинга JSON
        print(f"Ошибка при парсинге JSON: {e}")


# Запуск функции
if __name__ == "__main__":
    fetch_and_display_posts()
```
